```python
import matplotlib.pyplot as plt
import numpy as np
from ultralytics import YOLO
import random
import time
import cv2
```

```python
# ìƒ‰ìƒ íŒ”ë ˆíŠ¸
def random_color():
    return tuple([random.randint(0, 255) for _ in range(3)])
```

```python
connections = [(i, i + 1) for i in range(26)]  # 27ì  ê¸°ì¤€ (0~26)

for res in results :
    # ê°ì²´ë³„ë¡œ keypoints ì²˜ë¦¬
    for idx, result in enumerate(res):
        if result.keypoints is not None:
            keypts = result.keypoints.xy.cpu().numpy()
            for obj in keypts:
                color = random_color()
                # ì  ê·¸ë¦¬ê¸° + ë¼ë²¨
                for i, (x, y) in enumerate(obj):
                    if x > 0 and y > 0:
                        cv2.circle(result.orig_img, (int(x), int(y)), 5, color, -1)
                        cv2.putText(result.orig_img, str(i), (int(x), int(y) - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.4, color, 1)
                # ì„  ì—°ê²° (ì¸ë±ìŠ¤ê°€ ìœ íš¨í•œ ì ë§Œ ì—°ê²°)
                for i, j in connections:
                    if i < len(obj) and j < len(obj):  # ìœ íš¨í•œ ì ë§Œ ì—°ê²°
                        if all(obj[[i, j]].flatten() > 0):  # ìœ íš¨í•œ ì ë§Œ ì—°ê²°
                            pt1 = tuple(map(int, obj[i]))
                            pt2 = tuple(map(int, obj[j]))
                            cv2.line(result.orig_img, pt1, pt2, color, 2)

    # ê°ì²´ ê°ì§€ ì •ë³´ ì¶œë ¥
    for idx, box in enumerate(res.boxes):  # results[0]ë¡œ ì²« ë²ˆì§¸ ì´ë¯¸ì§€ì˜ ê²°ê³¼ ê°€ì ¸ì˜¤ê¸°
        if (idx == 0) :
            print("=="*20)

        xyxy = box.xyxy.cpu().numpy()  # ë°”ìš´ë”© ë°•ìŠ¤ ì¢Œí‘œ
        cls = box.cls.cpu().numpy()    # í´ëž˜ìŠ¤ ID
        conf = box.conf.cpu().numpy()  # confidence
        class_name = res[0].names[int(cls)]  # í´ëž˜ìŠ¤ ì´ë¦„ (results[0]ë¡œ ì²« ë²ˆì§¸ ê²°ê³¼ ì ‘ê·¼)
        # ì—¬ê¸°ì„œ confëŠ” ë°°ì—´ì´ë¯€ë¡œ, ì²« ë²ˆì§¸ ê°’ë§Œ ì‚¬ìš©
        conf_value = conf[0].item()  # .item()ìœ¼ë¡œ ìŠ¤ì¹¼ë¼ ê°’ ì¶”ì¶œ
        print(f"Object {idx}: Class {class_name} (ID: {cls}), Confidence: {conf_value:.2f}, XYXY: {xyxy}")
        # ë°”ìš´ë”© ë°•ìŠ¤ ê·¸ë¦¬ê¸° + í´ëž˜ìŠ¤ ì´ë¦„ ë° confidence í‘œì‹œ
        cv2.rectangle(result.orig_img,
                    (int(xyxy[0][0]), int(xyxy[0][1])),  # .item() ì‚¬ìš©í•˜ì§€ ì•Šê³  ì§ì ‘ ì¶”ì¶œ
                    (int(xyxy[0][2]), int(xyxy[0][3])),
                    (0, 255, 0), 2)
        cv2.putText(result.orig_img, f"{class_name} {conf_value:.2f}",
                    (int(xyxy[0][0]), int(xyxy[0][1]) - 10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 2) 
```

```python
# keypoints ì •ë³´ ì¶œë ¥
for result in results:
    if result.keypoints is not None:
        keypoints = result.keypoints.xy.cpu().numpy()
        confidences = result.keypoints.conf.cpu().numpy()  # ì‹ ë¢°ë„
        for i, (x, y) in enumerate(keypoints[0]):
            print(f"Keypoint {i}: (x: {x}, y: {y}), Confidence: {confidences[0][i]:.2f}")
    print("===" * 30)

# ì˜ˆì¸¡ ì†ë„ ì¶œë ¥
inference_time = end_time - start_time
print(f"Inference Time: {inference_time:.2f} seconds")
```

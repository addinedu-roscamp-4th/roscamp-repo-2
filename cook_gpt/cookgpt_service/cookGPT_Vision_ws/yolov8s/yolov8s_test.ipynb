```python
import matplotlib.pyplot as plt
import numpy as np
from ultralytics import YOLO
import random
import time
import cv2
```

```python
start_time = time.time()

model = YOLO('/home/addinedu/Desktop/Yolo_ws/yolov8s/runs/pose/train2/weights/best.pt')

img_dir_path = '/home/addinedu/Desktop/Yolo_ws/yolov8s/StickerPoseDetection/test/images'

results = model.predict(
    source=img_dir_path, 
)

end_time = time.time()

# 예측 속도 출력
inference_time = end_time - start_time
```

```python
# 색상 팔레트
def random_color():
    return tuple([random.randint(0, 255) for _ in range(3)])
```

```python
connections = [(i, i + 1) for i in range(26)]  # 27점 기준 (0~26)

for res in results :
    # 객체별로 keypoints 처리
    for idx, result in enumerate(res):
        if result.keypoints is not None:
            keypts = result.keypoints.xy.cpu().numpy()
            for obj in keypts:
                color = random_color()
                # 점 그리기 + 라벨
                for i, (x, y) in enumerate(obj):
                    if x > 0 and y > 0:
                        cv2.circle(result.orig_img, (int(x), int(y)), 5, color, -1)
                        cv2.putText(result.orig_img, str(i), (int(x), int(y) - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.4, color, 1)
                # 선 연결 (인덱스가 유효한 점만 연결)
                for i, j in connections:
                    if i < len(obj) and j < len(obj):  # 유효한 점만 연결
                        if all(obj[[i, j]].flatten() > 0):  # 유효한 점만 연결
                            pt1 = tuple(map(int, obj[i]))
                            pt2 = tuple(map(int, obj[j]))
                            cv2.line(result.orig_img, pt1, pt2, color, 2)

    # 객체 감지 정보 출력
    for idx, box in enumerate(res.boxes):  # results[0]로 첫 번째 이미지의 결과 가져오기
        if (idx == 0) :
            print("=="*20)

        xyxy = box.xyxy.cpu().numpy()  # 바운딩 박스 좌표
        cls = box.cls.cpu().numpy()    # 클래스 ID
        conf = box.conf.cpu().numpy()  # confidence
        class_name = res[0].names[int(cls)]  # 클래스 이름 (results[0]로 첫 번째 결과 접근)
        # 여기서 conf는 배열이므로, 첫 번째 값만 사용
        conf_value = conf[0].item()  # .item()으로 스칼라 값 추출
        print(f"Object {idx}: Class {class_name} (ID: {cls}), Confidence: {conf_value:.2f}, XYXY: {xyxy}")
        # 바운딩 박스 그리기 + 클래스 이름 및 confidence 표시
        cv2.rectangle(result.orig_img,
                    (int(xyxy[0][0]), int(xyxy[0][1])),  # .item() 사용하지 않고 직접 추출
                    (int(xyxy[0][2]), int(xyxy[0][3])),
                    (0, 255, 0), 2)
        cv2.putText(result.orig_img, f"{class_name} {conf_value:.2f}",
                    (int(xyxy[0][0]), int(xyxy[0][1]) - 10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 2) 
```

```python
# keypoints 정보 출력
for result in results:
    if result.keypoints is not None:
        keypoints = result.keypoints.xy.cpu().numpy()
        confidences = result.keypoints.conf.cpu().numpy()  # 신뢰도
        for i, (x, y) in enumerate(keypoints[0]):
            print(f"Keypoint {i}: (x: {x}, y: {y}), Confidence: {confidences[0][i]:.2f}")
    print("===" * 30)

# 예측 속도 출력
inference_time = end_time - start_time
print(f"Inference Time: {inference_time:.2f} seconds")
```

```python
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np

# 예시: results가 있다고 가정
# results = [...]

# subplot 사이즈 설정
n_rows = 8
n_cols = 6
fig, axes = plt.subplots(n_rows, n_cols, figsize=(n_cols * 3, n_rows * 3))  # figsize는 적절히 조절 가능

# 이미지 시각화
for idx, result in enumerate(results):
    if idx >= n_rows * n_cols:
        break  # 총 48개까지만 표시

    img_array = result.plot()  # 예측된 값을 BGR numpy 배열로 시각화
    img = Image.fromarray(img_array[..., ::-1])  # RGB PIL 이미지로 변환
    ax = axes[idx // n_cols, idx % n_cols]  # subplot 위치
    ax.imshow(img)
    ax.axis('off')  # 축 제거

# 여분 subplot 비우기
total_plots = n_rows * n_cols
if len(results) < total_plots:
    for idx in range(len(results), total_plots):
        axes[idx // n_cols, idx % n_cols].axis('off')

plt.tight_layout()
plt.show()
```
